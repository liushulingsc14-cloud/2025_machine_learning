```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def func(x):
    y = 1/(1+25*x**2)
    return y

x_values = np.linspace(-1, 1, 100)
y_values = func(x_values)

X = x_values.reshape(-1, 1)
Y = y_values


poly1 = PolynomialFeatures(degree=2)
poly2 = PolynomialFeatures(degree=5)
poly3 = PolynomialFeatures(degree=30)
poly4 = PolynomialFeatures(degree=10**5)

# 轉換 X 成多項式特徵
X_poly1 = poly1.fit_transform(X)
X_poly2 = poly2.fit_transform(X)
X_poly3 = poly3.fit_transform(X)
X_poly4 = poly4.fit_transform(X)

# 建立線性回歸模型並訓練
model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()
model4 = LinearRegression()

model1.fit(X_poly1, Y)
model2.fit(X_poly2, Y)
model3.fit(X_poly3, Y)
model4.fit(X_poly4, Y)

# 用模型做預測
y_pred1 = model1.predict(X_poly1)
y_pred2 = model2.predict(X_poly2)
y_pred3 = model3.predict(X_poly3)
y_pred4 = model4.predict(X_poly4)


plt.figure(figsize=(8, 5))
plt.plot(x_values, y_pred1, label=f"Polynomial deg={2}", color="green")
plt.plot(x_values, y_pred2, label=f"Polynomial deg={5}", color="purple")
plt.plot(x_values, y_pred3, label=f"Polynomial deg={30}", color="black", linewidth=3, linestyle='--')
plt.plot(x_values, y_pred4, label=f"Polynomial deg={100}", color="red")
plt.scatter(x_values, y_values, s=10, color="blue", alpha=0.5, label="Data points", linewidth=1)
plt.legend()
plt.show()
```
